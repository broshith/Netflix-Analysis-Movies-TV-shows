With Colab you can harness the full power of popular Python libraries to analyze and visualize data. The code cell below uses **numpy** to generate some random data, and uses **matplotlib** to visualize it. To edit the code, just click the cell and start editing.

You can import your own data into Colab notebooks from your Google Drive account, including from spreadsheets, as well as from Github and many other sources. To learn more about importing data, and how Colab can be used for data science, see the links below under [Working with Data](#working-with-data).

````
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
````

**Exploratory Data Analysis (EDA)**

**1. Number of movies vs TV shows**

````
from matplotlib import pyplot as plt
import seaborn as sns

sns.countplot(x='type', data=df,hue='type')
plt.show()
````

**2. Top 10 countries producing most content**

````
## Illa endhuku chesam ante duplicates remove chesthu vunde time lo manaki country column poyindi so illa chesi country column without duplicates or null values tho techham malli

import pandas as pd

df = pd.read_csv("netflix_titles_all.csv")

if 'country' not in df.columns:
    df['country'] = df['listed_in'].str.split(',').str[-1] ##Listed_in enti ante already vunna columns lo country name vuntadi kada daani use chesukoni teesukuntam str[-1] enti ante last element of list country name vuntadi adhi teesukuntam
df['country'] = df['country'].fillna('Unknown')   ## fillna ante konni areas lo columns lo country name lekunte unknown ane word tho fill cheayyamani maeaning
country_counts = df['country'].value_counts().head(10)

plt.figure(figsize=(10,6))
plt.plot(country_counts.index,country_counts.values,marker='o',color='b')
plt.xlabel("countries")
plt.ylabel("number of movies/tv shows")
plt.title("Top 10 countries producing most content")
plt.xticks(rotation=50)
print(country_counts)
````

**3. Most common genres**

````
import pandas as pd

df = pd.read_csv("netflix_titles_all.csv")

df['listed_in'] = df['listed_in'].str.split(',')
genre_counts = df.explode('listed_in')['listed_in'].value_counts().head(10) ##here we use explode function because in the listed_in each contains 3 to 4 genres so to take and create a each genre into single column we take explode this explode uses this property

plt.figure(figsize=(10,6))
plt.plot(genre_counts.index, genre_counts.values, marker='o', color='b')
plt.title('Most Common Genres on Netflix')
plt.xlabel('Genre')
plt.ylabel('Number of Movies/TV Shows')
plt.xticks(rotation=60)
plt.show()
print(genre_counts)
````

**4. Content added per year**

````
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("netflix_titles_all.csv")

df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce')
content_per_year = df['release_year'].value_counts().sort_index()

plt.figure(figsize=(10,6))
plt.plot(content_per_year.index, content_per_year.values, marker='o', color='b')
plt.title('Content Added per Year on Netflix')
plt.xlabel('Year')
plt.ylabel('Number of Movies/TV Shows')
plt.show()
print(content_per_year)
````

**5. Ratings distribution**

````
import pandas as pd
from matplotlib import pyplot as plt

df = pd.read_csv("netflix_titles_all.csv")

df['rating'] = df['rating'].fillna("unknown")  ## Ikkada endhuku illa chesam ante dropna() function pina use chesinappudu poyindi ratings column
rating_counts = df['rating'].value_counts()

plt.figure(figsize=(10,6))
plt.plot(rating_counts.index,rating_counts.values,marker='o',color='b')
plt.title("Rating Distribution")
plt.xlabel("Ratings")
plt.ylabel("Number of Movies/TV Shows on Netflix")
plt.xticks(rotation=60)
plt.show()
print(rating_counts)
````

**4. Heatmaps for missing values**

````
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

df = pd.read_csv("netflix_titles_all.csv")
missing_values = df.isnull()

plt.figure(figsize=(12,8))
sns.heatmap(missing_values, cbar=False, cmap='viridis', annot=False)
plt.title("Missing Values Heatmap")
plt.show()
````

**5. Content Distribution by Rating**

````
import pandas as pd
from matplotlib import pyplot as plt

df['rating'] = df['rating'].fillna("unknown")
rating_counts = df['rating'].value_counts()

plt.figure(figsize=(12,8))
plt.bar(rating_counts.index,rating_counts.values,color="b")
plt.xlabel("Ratings")
plt.ylabel("Movies/TV Shows on Netflix")
plt.title("Content Distribution by Rating")
plt.show()
````
